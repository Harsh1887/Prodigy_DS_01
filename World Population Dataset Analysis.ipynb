<b><h3> 5. Visualizing Total Population</h3></b>

A. Total Population of Top 10 Countries

```python
total_population_data = population[population['Series Code'] == 'SP.POP.TOTL']
total_population_sorted = total_population_data.sort_values(by='2022', ascending=False)
total_top_ten_countries = total_population_sorted.head(10)
```

Bar Plot - Top 10(2022)

```python
plt.figure(figsize=(10,6))
sns.barplot(x='2022', y='Country Name', hue='2022', legend=False, data=total_top_ten_countries, palette='coolwarm')
plt.title('Top 10 Countries by Total Population (2022)')
plt.xlabel('Total Population')
plt.ylabel('Country')
plt.show()
```

<b><h4>Historical Population Trends (2016,2010,2001,) of the Top 10 Most Populous Contries in 2022

```python
for year in ['2005', '2010', '2001']:
    plt.figure(figsize=(10,6))
    sns.barplot(x=year, y='Country Name', hue=year, legend=False, data=total_top_ten_countries, palette='viridis')
    plt.title(f'Historical Population Trends ({year}) of the Top 10 Most Populous Countries in 2022')
    plt.xlabel('Total Population')
    plt.ylabel('Country')
    plt.show()
```

<b><h4>B. Bottom 10 Countries by Total Population (2022, 2016, 2010, 2001)

```python
total_botton_ten_countries = total_population_data.sort_values(by='2022', ascending=True).head(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='2022', y='Country Name', hue='2022', legend=False, data=total_botton_ten_countries, palette='magma')
plt.title('Bottom 10 Countries by Total Population (2022)')
plt.xlabel('Total Population')
plt.ylabel('Country')
plt.show()
```

#### Bar Plot – Historical Population Trends (2016, 2010, 2001) of the Bottom 10 Most Populous Countries in 2022


```python
for year in ['2006', '2010', '2001']:
    plt.figure(figsize=(10,6))
    sns.barplot(x=year, y='Country Name', hue=year, legend=False, data=total_botton_ten_countries, palette='magma')
    plt.title(f'Historical Population Trends ({year}) of the Bottom 10 Most Populous Countries in 2022')
    plt.xlabel('Total Population')
    plt.ylabel('Country')
    plt.show()
```

<b><h3>6. Male and Female Population</h3></b>

<h4>A. Top and Bottom Male Population Countries(2022)

```python
male_population_data = population[population['Series Code'] == 'SP.POP.TOTL.MA.IN']
male_population_sorted = total_population_data.sort_values(by='2022', ascending=False)
male_top_ten_countries = total_population_sorted.head(10)
male_bottom_ten_countries = total_population_sorted.tail(10)
```

#### Top 10 Male Population Countries(2022)

```python
plt.figure(figsize=(10,6))
sns.barplot(x='2022', y='Country Name', hue='2022', legend=False, data=male_top_ten_countries, palette='crest')
plt.title('Bottom 10 Countries by Male Population (2022)')
plt.xlabel('Total Population')
plt.ylabel('Country')
plt.show()
```

#### Botttom 10 Male Poplation Countries(2022)

```python
plt.figure(figsize=(10,6))
sns.barplot(x='2022', y='Country Name', hue='2022', legend=False, data=male_bottom_ten_countries, palette='crest')
plt.title('Bottom 10 Countries by Male Population (2022)')
plt.xlabel('Total Population')
plt.ylabel('Country')
plt.show()
```

### B. Top and Bottom 10 Female Population Countries(2022)

```python
female_population_data = population[population['Series Code'] == 'SP.POP.TOTL.MA.IN']
female_population_sorted = total_population_data.sort_values(by='2022', ascending=False)
female_top_ten_countries = total_population_sorted.head(10)
female_bottom_ten_countries = total_population_sorted.tail(10)
```

#### Top 10 Female Population Countries(2022)

```python
plt.figure(figsize=(10,6))
sns.barplot(x='2022', y='Country Name', hue='2022', legend=False, data=female_top_ten_countries, palette='magma')
plt.title('Bottom 10 Countries by Female Population (2022)')
plt.xlabel('Total Population')
plt.ylabel('Country')
plt.show()
```

#### Botttom 10 Female Poplation Countries(2022)

```python
plt.figure(figsize=(10,6))
sns.barplot(x='2022', y='Country Name', hue='2022', legend=False, data=female_bottom_ten_countries, palette='magma')
plt.title('Bottom 10 Countries by Female Population (2022)')
plt.xlabel('Total Population')
plt.ylabel('Country')
plt.show()
```

<b><h3>7. Stacked Bar Plot for Male and Female Population

```python
merged_data = pd.merge(male_population_data, female_population_data, on='Country Name', suffixes=('_male', '_female'))
merged_data['Total Population'] = merged_data['2022_male'] + merged_data['2022_female']
```

```python
top_10_countries = merged_data.sort_values(by='Total Population', ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Country Name', y='2022_female', data=top_10_countries, color='pink', label='Female')
sns.barplot(x='Country Name', y='2022_male', data=top_10_countries, bottom=top_10_countries['2022_female'], color='blue', label='Male')
plt.title('Bottom 10 Countries by Female Population (2022)')
plt.xlabel('Total Population')
plt.ylabel('Country')
plt.show()
```

```python
bottom_10_countries = merged_data.sort_values(by='Total Population', ascending=True).head(10)
```

```python
plt.figure(figsize=(12, 6))

# Bar positions
countries = bottom_10_countries['Country Name']
x = range(len(countries))

# Female bars
female = bottom_10_countries['2022_female']
male = bottom_10_countries['2022_male']

# Plot female bars
plt.bar(x, female, color='pink', label='Female')

# Plot male bars stacked on top of female
plt.bar(x, male, bottom=female, color='blue', label='Male')

# Labels and titles
plt.xticks(ticks=x, labels=countries, rotation=45, ha='right')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Bottom 10 Countries by Female and Male Population (2022)')
plt.legend()
plt.tight_layout()
plt.show()
```
